```ocaml
let error f = Printf.ksprintf (fun s -> Firebug.console##error (Js.string s); failwith s) f

let original_canvas =   Js.Opt.get
        (Js.Opt.bind ( Dom_html.document##getElementById("canvas_original") )
            Dom_html.CoerceTo.canvas)
        (fun () -> error "can't find canvas element original")

let image : Dom_html.imageElement Js.t = Dom_html.createImg Dom_html.document;;

let c2D_orig = original_canvas##getContext (Dom_html._2d_);;

let _ = 
    image##src <- Js.string "../tutos/lena.png";
    c2D_orig##drawImage (image, 0., 0.);;
```

```ocaml
let original_imageData = c2D_orig##getImageData (0., 0., 512., 512.)
  
let original_data = original_imageData##data

let final_canvas =   Js.Opt.get
        (Js.Opt.bind ( Dom_html.document##getElementById("canvas_final") )
            Dom_html.CoerceTo.canvas)
        (fun () -> error "can't find canvas element final")

let image2 : Dom_html.imageElement Js.t = Dom_html.createImg Dom_html.document

let c2D_final = final_canvas##getContext (Dom_html._2d_)
    
let final_imageData = c2D_final##getImageData (0., 0., 512., 512.)
let final_data = final_imageData##data
```

```ocaml
let imgData_to_vector data = 
    let vect = Vector.create Vector.int32 (512*512*4) in
    for i = 0 to Vector.length vect - 1 do
    vect.[<i>] <- Int32.of_int (Dom_html.pixel_get data i);
  done;
  vect

let vector_to_imgData vect data = 
    for i = 0 to Vector.length vect - 1 do
    Dom_html.pixel_set data i (Int32.to_int vect.[<i>])
  done
```

*TODO*

```ocaml
(* default : color -> gray *)
let filter = kern v ->
  let open Std in
  let tid = thread_idx_x + block_dim_x * block_idx_x in
  if tid <= (512*512) then (
    let i = (tid*4) in
    let res = int_of_float ((0.21 *. (float (v.[<i>]))) +.
                           (0.71 *. (float (v.[<i+1>]))) +.
                           (0.07 *. (float (v.[<i+2>]))) ) in
    v.[<i>] <- res;
    v.[<i+1>] <- res;
    v.[<i+2>] <- res )
    
let _ = 
  ignore(Kirc.gen ~only:Devices.OpenCL filter);
```

*TODO*

```ocaml
let dev = devs.(0)
let threadsPerBlock = match dev.Devices.specific_info with
    | Devices.OpenCLInfo clI -> 
      (match clI.Devices.device_type with
       | Devices.CL_DEVICE_TYPE_CPU -> 1
       | _  ->   256)
    | _  -> 256 
  let blocksPerGrid =
    ((512*512) + threadsPerBlock -1) / threadsPerBlock
  
  let block0 = {Spoc.Kernel.blockX = threadsPerBlock;
      Spoc.Kernel.blockY = 1; Spoc.Kernel.blockZ = 1}
      
  let grid0= {Spoc.Kernel.gridX = blocksPerGrid;
        Spoc.Kernel.gridY = 1; Spoc.Kernel.gridZ = 1}
  
```

*TODO*

```ocaml
let vect = imgData_to_vector original_data

let run () =
    Kirc.run filter vect (block0,grid0) 0 dev;
    vector_to_imgData vect final_data;
    c2D_final##putImageData (final_imageData, 0., 0.);
```

```ocaml
let _ = 
    run ()
```
